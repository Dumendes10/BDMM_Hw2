```python
from pprint import pprint

pipeline = [
    # Filter listings with superhost and capacity > 4
    {
        "$match": {
            "host_is_superhost": True,
            "accommodates": { "$gt": 4 }
        }
    },

    # Group by host and count qualifying listings
    {
        "$group": {
            "_id": "$host_id",
            "host_name": { "$first": "$host_name" },
            "total_qualifying": { "$sum": 1 },
            "listings": {
                "$push": {
                    "name": "$name",
                    "capacity": "$accommodates"
                }
            }
        }
    },

    # Only include hosts with 2 or more qualifying listings
    {
        "$match": {
            "total_qualifying": { "$gte": 2 }
        }
    },

    # Random sample of 3 superhosts
    {
        "$sample": { "size": 3 }
    }
]

# Run and print results
results = db.listingsAndReviews_new.aggregate(pipeline)
for result in results:
    pprint(result)
```

```python
pipeline = [
    # Listings in New York with reviews
    {
        "$match": {
            "address.market": "New York",
            "reviews.0": { "$exists": True }
        }
    },

    # Unwind reviews array
    { "$unwind": "$reviews" },

    # Step 3: Filter reviews that actually have a comment
    {
        "$match": {
            "reviews.comments": { "$type": "string" }
        }
    },

    # Project reviewer name, review, and string length
    {
        "$project": {
            "_id": 0,
            "reviewer": "$reviews.reviewer_name",
            "comment": "$reviews.comments",
            "length": { "$strLenCP": "$reviews.comments" },
            "listing": "$name"
        }
    },

    # Sort by length descending
    { "$sort": { "length": -1 } },

    # Return top 1
    { "$limit": 1 }
]

results = db.listingsAndReviews_new.aggregate(pipeline)

for result in results:
    pprint(result['reviewer'])
```

### Question 5

```python
pipeline = [
    # Only listings with required fields
    {
        "$match": {
            "price": { "$exists": True },
            "security_deposit": { "$exists": True },
            "accommodates": { "$exists": True, "$gt": 0 }
        }
    },

    # Project absolute difference between price and deposit
    {
        "$project": {
            "_id": 0,
            "accommodates": 1,
            "diff": {
                "$abs": { "$subtract": ["$price", "$security_deposit"] }
            }
        }
    },

    # Group by accommodates and get min/max difference
    {
        "$group": {
            "_id": "$accommodates",
            "max_diff": { "$max": "$diff" },
            "min_diff": { "$min": "$diff" }
        }
    },

    # Sort by accommodates
    { "$sort": { "_id": 1 } }
]

# Run and show results
results = db.listingsAndReviews_new.aggregate(pipeline)
for result in results:
    pprint(result)
```

```python
query_1 = {
    "$project": {
        "name": 1,
        "security_score": {
            "$divide": [
                {
                    "$subtract": ["$price", {"$ifNull": ["$security_deposit", 0]}]
                },
                "$guests_included"
            ]
        }
    }
}

query_2 = {'$sort': {'security_score': -1}}

query_3 = {"$group": {
        "_id": None,
        "most_secure": {"$first": "$$ROOT"},
        "least_secure": {"$last": "$$ROOT"}
    }}

pipeline = [query_1, query_2, query_3]

result = db.listings.aggregate(pipeline)

for i in result:
    pprint(i)
```

```python
pipeline = [
    {
        "$project": {
            "listing_id": "$_id",
            "price": { "$toDouble": "$price" },  # convert Decimal128 to number
            "available_days": "$availability.availability_30",  # or _60/_90/_365
            "total_transaction_value": {
                "$multiply": [
                    { "$toDouble": "$price" },
                    "$availability.availability_30"
                ]
            },
            "calendar_last_scraped": 1
        }
    },
    {
        "$project": {
            "listing_id": 1,
            "average_transaction_value": {
                "$cond": [
                    { "$gt": ["$available_days", 0] },
                    { "$divide": ["$total_transaction_value", "$available_days"] },
                    None
                ]
            },
            "available_days": 1,
            "calendar_last_scraped": 1
        }
    },
    { "$limit": 5 }
]

results = list(db.listingsAndReviews_new.aggregate(pipeline))
for r in results:
    pprint(r)
```

```python
# If you need to support fast filtering by listing_id
db.listingsAndReviews_new.create_index([("_id", 1)])
```

This model assumes a fixed price from the price field. If we had daily prices (like in a real calendar), we could sum those directly instead.
We estimated average transaction value using price*availability_30, since no actual transaction collection exists. The result is calculated using $project, $multiply, and $divide, and listings with no availability return None. An index on listing ID ensures fast search access.

### Question 11

```python
pipeline_city_stats = [
    { "$match": { "address.market": { "$exists": True }, "price": { "$exists": True } } },
    {
        "$group": {
            "_id": {
                "city": "$address.market",
                "property_type": "$property_type"
            },
            "count": { "$sum": 1 },
            "avg_price": { "$avg": { "$toDouble": "$price" } }
        }
    },
    {
        "$group": {
            "_id": "$_id.city",
            "property_stats": {
                "$push": {
                    "property_type": "$_id.property_type",
                    "count": "$count",
                    "avg_price": "$avg_price"
                }
            },
            "total_properties": { "$sum": "$count" }
        }
    },
    { "$sort": { "total_properties": -1 } },
    { "$limit": 10 },
    { "$merge": { "into": "top_destinations", "on": "_id", "whenMatched": "replace" } }
]

db.listingsAndReviews_new.aggregate(pipeline_city_stats)
```

```python
top_cities = db.top_destinations.find()
for city_doc in top_cities:
    city = city_doc["_id"]
    top_listings = list(db.listingsAndReviews_new.find(
        { "address.market": city, "price": { "$exists": True } },
        {
            "name": 1,
            "price": 1,
            "number_of_reviews": 1,
            "review_scores_rating": 1
        }
    ).sort("number_of_reviews", -1).limit(3))

    db.top_destinations.update_one(
        { "_id": city },
        { "$set": { "top_listings": top_listings } }
    )
```
